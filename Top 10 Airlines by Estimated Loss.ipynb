```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')


### Bar graph
#Top 10 airlines by estimated loss

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])
top_10 = df.nlargest(10,'estimated_loss_usd')

#airline_counts = df[top_10['airline']].value_counts()
#plt.figure()
plt.figure(figsize = (10,7))
plt.barh(top_10['airline'],top_10['estimated_loss_usd'])
plt.title('Top 10 Airlines by Estimated Loss')

plt.xlabel('Estimated loss (in Millions USD)')
plt.ylabel('Airlines')

plt.gca().invert_yaxis() 
plt.tight_layout()
plt.savefig('Top_10_Airlines_by_estimated_loss_barh.png')
plt.show() 

```
