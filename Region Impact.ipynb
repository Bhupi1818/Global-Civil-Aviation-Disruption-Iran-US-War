```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')


### Bar graph
# region and estimated_loss_usd

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])
#top_10 = df.nlargest(10,'estimated_loss_usd')

#airline_counts = df[top_10['airline']].value_counts()
#plt.figure()
plt.figure(figsize = (8,6))
plt.bar(df['region'],df['estimated_loss_usd'])
plt.title('Region Impact')

plt.xlabel('Region')
plt.ylabel('Estimated_Loss_USD (in Millions)')

#plt.gca().invert_yaxis() 
plt.tight_layout()
plt.savefig('Region_and_estimated_loss_usd.png')
plt.show() 

```
