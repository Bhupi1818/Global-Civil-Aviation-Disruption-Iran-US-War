```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

# Cancellation VS Reroute

top_10 = df.nlargest(10, 'estimated_loss_usd')

airlines = top_10['airline']
cancellations = top_10['cancellations_count']
reroutes = top_10['reroutes_count']

x = np.arange(len(airlines))  
width = 0.35                  

plt.figure(figsize=(14, 7))

plt.bar(x - width/2, cancellations, width, label='Cancellations', color='tomato')

plt.bar(x + width/2, reroutes, width, label='Reroutes', color='dodgerblue')

plt.title('Cancellations VS Reroutes for Top 10 Most Affected Airlines')
plt.xlabel('Airlines')
plt.ylabel('Number of Flights')


plt.xticks(x, airlines, rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.savefig('Cancellations_VS_Reroutes_for_Top_10_Most_Affected_Airlines.png')
plt.show()
```
