```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

# Comparision Between Cancellation VS Reroutes

cancellations_data = df.groupby(['cancellations_count','airline_type']).size().unstack().fillna(0)
reroutes_data = df.groupby(['reroutes_count','airline_type']).size().unstack().fillna(0)
fig,ax = plt.subplots(1,2,figsize = (12,6))

#first subplot:Cancellations_count
cancellations_data.plot(ax = ax[0])
ax[0].set_title('Cancelled_Airline')
ax[0].set_xlabel('Cancellations Count')
ax[0].set_ylabel('Frequency')

#second subplot:reroutes_count
reroutes_data.plot(ax = ax[1])
ax[1].set_title('Reroute_Airline')
ax[1].set_xlabel('Reroutes Count')
ax[1].set_ylabel('Frequency')

fig.suptitle('Comparision Of Cancellations and Reroutes')

#plt.tight_layout()
plt.savefig('Comparision_Of_Cancellationa_and_Reroutes.png')
plt.show()
```
