```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

# Total Financial Loss by Region
# Pie Chart
top_10_airlines = df.nlargest(10,'revenue_loss_pct')

plt.figure(figsize=(10,8))
plt.pie(top_10_airlines['revenue_loss_pct'].values, labels=top_10_airlines['airline'], autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Airlines by Revenue Loss Percentage')
plt.tight_layout()
plt.savefig('Top_10_Airlines_by_REvenue_loss_Percentage_pie.png')
plt.show()
```
