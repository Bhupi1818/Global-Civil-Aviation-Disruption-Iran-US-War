```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

#Cancellations VS reroutes
#Scatter plot

plt.figure(figsize=(8,4))
plt.scatter(df['reroutes_count'],df['cancellations_count'],color='blue')
plt.title('Cancellations VS Reroutes')
plt.xlabel('Number of Reroutes')
plt.ylabel('Number of Cancellations')
plt.tight_layout()
plt.savefig('Cancellations_VS_Reroutes_scatter.png')
plt.show() 

```
