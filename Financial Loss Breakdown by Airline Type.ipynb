```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

# Total Financial Loss by airline_type
# Pie Chart
airline_type_losses = df.groupby('airline_type')['revenue_loss_pct'].sum()

plt.figure(figsize=(20,16))
plt.pie(airline_type_losses.values, labels=airline_type_losses.index, autopct='%1.1f%%', startangle=90)
plt.title('Financial Loss Breakdown by Airline Type')
plt.tight_layout()
plt.savefig('Financial_LossBreakdown_by_Airline_Type_pie.png')
plt.show()
```
