```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')
df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

# Line graph
# Airline and Revenue_loss_pct

top_10_pct = df.nlargest(10,'revenue_loss_pct')
top_10_pct = top_10_pct.sort_values('revenue_loss_pct', ascending=True)

plt.figure(figsize=(10,6))

plt.hlines(y=top_10_pct['airline'],xmin=0,xmax=top_10_pct['revenue_loss_pct'])
plt.plot(top_10_pct['revenue_loss_pct'],top_10_pct['airline'],marker='o')

plt.title('Top 10 Airlines by Revenue Loss Percentage')
plt.xlabel('Revenue Lost (%)')
plt.ylabel('Airline')
plt.grid(color='blue',linestyle='--')

plt.tight_layout()
plt.savefig('Top_10_Airlines_by_Revenue_Loss_pct_lollipop.png', dpi=300)
plt.show()
```
