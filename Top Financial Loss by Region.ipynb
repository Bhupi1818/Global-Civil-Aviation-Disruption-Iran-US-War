```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df = pd.read_csv('/Users/bhupendrasaini/Downloads/airline_losses.csv')

df = df.dropna(subset = ['airline','country','airline_type','cancellations_count','reroutes_count','revenue_loss_pct','estimated_loss_usd','region'])

# Total Financial Loss by Region
# Pie Chart
region_losses = df.groupby('region')['estimated_loss_usd'].sum()

plt.figure(figsize=(20,16))
plt.pie(region_losses.values, labels=region_losses.index, autopct='%1.1f%%', startangle=90)
plt.title('Top Financial Loss by Region')
plt.tight_layout()
plt.savefig('Top_Financial_Loss_by_Region_pie.png')
plt.show()
```
